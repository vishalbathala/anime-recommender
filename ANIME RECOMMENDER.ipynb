```python
#importing libraries
import pandas as pd
import numpy as np
import scipy as sp
from sklearn.metrics.pairwise import cosine_similarity
```

```python
anime_df = pd.read_csv('anime.csv')
anime_df
```

```python
rating_df = pd.read_csv('rating.csv')
rating_df
```

```python
anime_df.info()
```

```python
anime_df.shape
```

```python
rating_df.info()
```

```python
type_mode = anime_df['type'].mode()
```

```python
genre_mode = anime_df['genre'].mode()
```

```python

#deleting anime with 0 ratings
anime_df = anime_df[~np.isnan(anime_df['rating'])]
```

```python
anime_df.info()
```

```python
#filling mode values for genre and type

anime_df['genre'] = anime_df['genre'].fillna(genre_mode[0])
anime_df['type'] = anime_df['type'].fillna(type_mode[0])
```

```python
# checking whether all null values filled or not
anime_df.isnull().sum()
```

```python
# in ratings -1 suggests that user has not given rating so we will fill with NaN 
rating_df['rating'] = rating_df['rating'].apply(lambda x:np.nan if x==-1 else x)
```

```python
rating_df.head(20)
```

```python
#now we have to recommend anime series which is of type TV only
anime_df = anime_df[anime_df['type']=='TV']
```

```python
anime_df
```

```python
#We make a new Dataframe combining both anime and rating on the anime_id column.
rated_anime_df = rating_df.merge(anime_df, left_on = 'anime_id', right_on = 'anime_id', suffixes= ['_user', ''])
```

```python

rated_anime_df
```

```python
#Leaving only user_id, name and rating as the Df.
rated_anime_df = rated_anime_df[['user_id','name','rating']]
rated_anime_df
```

```python
#we will compute this based on first 7500 users only
rated_anime_first_7500_df = rated_anime_df[rated_anime_df.user_id<=7500]
```

```python
rated_anime_first_7500_df
```

```python
#creating pivot table for calculation of similarity

pivot_df = rated_anime_first_7500_df.pivot_table(index = ['user_id'],columns = ['name'],values = 'rating')
pivot_df
```

```python
#normalization

pivot_n_df = pivot_df.apply(lambda x: (x-np.mean(x))/(np.max(x)-np.min(x)),axis=1)
pivot_n_df
```

```python
#filling NaN values with 0
pivot_n_df=pivot_n_df.replace(np.nan,0)
pivot_n_df
```

```python
#transposing pivot table for the next step
pivot_n_df = pivot_n_df.T
```

```python
pivot_n_df
```

```python
#dropping columns with values 0 rating 
pivot_n_df = pivot_n_df.loc[:, (pivot_n_df != 0).any(axis=0)]
```

```python
pivot_n_df
```

```python
#Using scipy package to convert to sparse matrix format for the similarity computation
pivot_sparse = sp.sparse.csr_matrix(pivot_n_df.values)
```

```python
#model based on anime similarity

anime_similarity = cosine_similarity(pivot_sparse)
```

```python
#Df of anime similarities
anime_sim_df = pd.DataFrame(anime_similarity,index = pivot_n_df.index,columns = pivot_n_df.index)
```

```python
def anime_recomendation(anime_name):
    number = 1
    print('recomended because you watched {}:\n'.format(anime_name))
    for anime in anime_sim_df.sort_values(by =anime_name , ascending = False).index[1:6]:
        print(f'#{number}: {anime}   -   {round(anime_sim_df[anime][anime_name]*100,2)} %match')
        number+=1
```

```python
anime_recomendation('Code Geass: Hangyaku no Lelouch')
```
